```python
products["Unit Cost USD"]=products ["Unit Cost USD"].str.strip("$")

products["Unit Cost USD"]=products["Unit Cost USD"].str.replace(',',"")

```

```python
products["Unit Cost USD"] = products["Unit Cost USD"].astype('float')
```

```python
products["Unit Price USD"] = products["Unit Price USD"].str.strip("$")

products["Unit Price USD"] = products["Unit Price USD"].str.replace(",","")

products["Unit Price USD"] = products["Unit Price USD"].astype(float)

products
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(products['Category'].dropna(), kde=True, bins=20, color='skyblue')

plt.title('Distribution of Category')

plt.xlabel('Category')

plt.ylabel('Frequency')

plt.show()
```

```python
# Count plot of app categories

plt.figure(figsize=(12, 6))

sns.countplot(y='Category', data=products, order = products['Category'].value_counts().index, palette='viridis')

plt.title('Count of Product Name')

plt.xlabel('Product Name')

plt.ylabel('Category')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Brand', data=products, color='skyblue')

# Set the title and labels
plt.title('Distribution of ')
plt.xlabel('Brand')

# Show the plot
plt.show()
```
