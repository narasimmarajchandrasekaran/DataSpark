```python
plt.figure(figsize=(10, 6))

sns.histplot(stores['Country'].dropna(), kde=True, bins=20, color='skyblue')

plt.title('Distribution of Country')

plt.xlabel('Country')

plt.ylabel('Frequency')

plt.show()
```

```python
# Count plot of app categories

plt.figure(figsize=(12, 6))

sns.countplot(y='Country', data=stores, order = stores['Country'].value_counts().index, palette='viridis')

plt.title('Count of Open Date')

plt.xlabel('Open Date')

plt.ylabel('Country')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', data=stores, color='skyblue')

# Set the title and labels
plt.title('Distribution of Stores')
plt.xlabel('Country')

# Show the plot
plt.show()
```
